# TerraE Land Model: Sample Run and Visualization

14-day bare soil simulation with diurnal forcing and precipitation events.

```python
import sys
from pathlib import Path

# Run notebook from repo root
ROOT = Path(".").resolve()
if (ROOT / "src" / "terrae").exists():
    sys.path.insert(0, str(ROOT / "src"))
if (ROOT / "scripts").exists():
    sys.path.insert(0, str(ROOT / "scripts"))
import numpy as np
import matplotlib.pyplot as plt

from terrae import types
from terrae.driver import advance_bare_soil, init_step
from terrae.soil.heat import retp
```

```python
# Run simulation (or load from output/)
from sample_run import run_simulation

data, meta = run_simulation(n_days=14, dt_hours=1.0)
time_h, theta, tp, w_tot, runoff_cum, pr, zc, dz = data
```

```python
# Quick overview
time_d = time_h / 24
print(f"Steps: {len(time_d)}")
print(f"Layers: {meta['n_layers']}")
print(f"Final water: {w_tot[-1]*1000:.1f} mm")
print(f"Total runoff: {runoff_cum[-1]*1000:.1f} mm")
```

```python
# Generate full visualization
from visualize_run import load_run, plot_all, plot_water_balance

data_dict = load_run()
plot_all(data_dict)
plot_water_balance(data_dict)
```

## Interactive plots

```python
fig, ax = plt.subplots(figsize=(10, 5))
for k in range(theta.shape[1]):
    ax.plot(time_d, theta[:, k], label=f'Layer {k+1}')
ax.set_xlabel('Time (days)')
ax.set_ylabel(r'$\theta$ (m³/m³)')
ax.legend()
ax.set_title('Soil moisture time series')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
im = ax.pcolormesh(time_d, -zc, theta.T, shading='auto', cmap='YlGnBu')
ax.set_xlabel('Time (days)')
ax.set_ylabel('Depth (m)')
ax.set_title('Soil moisture Hovmöller')
ax.invert_yaxis()
plt.colorbar(im, ax=ax, label=r'$\theta$ (m³/m³)')
plt.tight_layout()
plt.show()
```
